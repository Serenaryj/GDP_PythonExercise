```python
import pandas as pd
import numpy as np
import os
```

```python
#Set working directory and read data
path = '/Users/serena/Desktop/Projects/Excercise/Python Exercise GDP/data_files'
os.chdir(path)
data_gdp= pd.read_csv('GDP.csv', header =1)   #Read csv file as pandas dataframe
data_gdp.set_index('Country Code', inplace = True)    #Set 'Country Code' as index
data_gdp.duplicated().sum()        #Check for duplicates, if returns 0, no duplicate
data_gdp2 = data_gdp.drop(['Country Name', 'Indicator Name', 'Indicator Code'], axis = 1)  # Drop redundant string variables
wdata_gdp = data_gdp2.transpose()     # Change dataframe shape
wdata_gdp = wdata_gdp.dropna(how='all')   # Drop rows with all mising values
wdata_gdp
```

```python
# Identify countries in Sub-Saharan Africa
data_region = pd.read_csv('Metadata_Country.csv')      # Read csv file to pandas dataframe
data_region.set_index('Country Code', inplace = True)            # Set index as 'Country Code'
data_sub = data_region[data_region.Region == "Sub-Saharan Africa"]   # Keep only countries in Sub-Saharan Africa
data_sub
```

```python
# Calculate correlation coefficients for GDP in sub-Saharan countries
list_sub = data_sub.index.tolist()     # Obtain list of Sub-Saharan countries
wdata_gdp_sub = wdata_gdp[list_sub]      # Keep GDP for sub-Saharan countries
df_corr = wdata_gdp_sub.corr()            # Compute correlation table for GDP in sub-Saharan countries
df_corr
```

```python
# Identify which country occurs most frequently as asked in the question
df_corr.values[[np.arange(len(df_corr))]*2] = 0    #Set the diagnal to 0
df_corr['Max Corr'] = df_corr.max(axis=1)         # Obtain the maximum correlation for each country
df_corr['Max Corr Country Code'] = df_corr.idxmax(axis=1)    # Identify the country code for the maximum correlation 
freq = df_corr['Max Corr Country Code'].value_counts()   #Count the frequency of each -country code
freq.index[0]           # Obtain the country code which occurs most
```
